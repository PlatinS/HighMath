#### 1. Пример функции, не имеющей предела в нуле и в бесконечностях

$\lim\limits_{x\to+\infty} \left( \dfrac{|x|}{x} * 2^x \right)$  
$\lim\limits_{x\to-\infty} \left( \dfrac{|x|}{x} * 2^x \right)$

```python
plt.figure(figsize=(6,6))
x = np.linspace(-4, 4, 100)
y = (abs(x)/x) * (2**x)
plt.plot(x, y, color='green')
plt.plot([-4, 4], [0, 0], color='blue', linestyle='--')
plt.plot([0, 0], [-4, 4], color='blue', linestyle='--')
plt.axis([-4, 4, -4, 4])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

#### Тема “Теоремы о пределах”.   
Найти предел:

$\displaystyle \lim\limits_{x\to 0}\frac{sin(2x)}{4x}=\frac{1}{2}$

$\displaystyle \lim\limits_{x\to 0}\frac{x}{sin(x)}=1$

$\displaystyle \lim\limits_{x\to 0}\frac{x}{arcsin(x)}=|t=sin(x)|=\lim\limits_{x\to 0}\frac{sin(t)}{t}=\lim\limits_{x\to 0}\frac{\frac{sin(t)}{t}}{1}=\frac{1}{1}=1$

$\displaystyle \lim\limits_{x\to \infty}(\frac{4x+3}{4x-3})^{6x}=
\big(\frac{4x - 3 + 6}{4x - 3}\big)^{6x} = \big(1 + \frac{6}{4x - 3}\big)^{6x} =
\bigg(1 + \frac{1}{\frac{4x -3}{6}}\bigg)^{6x}=\Bigg(\bigg(1 + \frac{1}{\frac{4x -3}{6}}\bigg)^{\frac{4x -3}{6}} \Bigg)^{\frac{6x \cdot 6}{4x - 3}} = \displaystyle e^{\frac{36x}{4x - 3}} = e^9$

$\displaystyle \lim\limits_{x\to \infty}\frac{sin(x) \cdot ln(x)}{x}=\lim\limits_{x\to \infty}\frac{sin(x)}{x}+\lim\limits_{x\to \infty}\frac{ln(x)}{x}=0+0=0$
