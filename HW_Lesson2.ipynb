1.2 Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

```python
import numpy as np
```

```python
def len_vector(v):
    return np.sqrt(np.sum(np.power(v, 2)))
```

2. Почему прямые не кажутся перпендикулярными?

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2=(-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
```

Две прямые перпендикулярны, если произведение их угловых коэффициентов равно -1.  
Т.к. 3*(-1/3) = (-1), то прямые перпендикулярны. Они не кажутся перпендикулярными из-за разного масштаба осей.

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2=(-1/3)*x+1
plt.figure(figsize=[6, 6])
plt.plot(x,y)
plt.plot(x,y2)
plt.xlim([-8, 8])
plt.ylim([-8, 8])
plt.xlabel('x')
plt.ylabel('y')
```

3.3 График гиперболы

```python
# k > 0 - гипербола определена в I и III координатных четвертях
k = 4
x1 = np.linspace(-10, -.1, 100)
x2 = np.linspace(.1, 10, 100)
y1 = k / x1
y2 = k / x2
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
# k < 0 - гипербола определена во II и IV координатных четвертях
k = -2
x1 = np.linspace(-10, -.1, 100)
x2 = np.linspace(.1, 10, 100)
y1 = k / x1
y2 = k / x2
plt.plot(x1,y1)
plt.plot(x2,y2)
```

5.1 Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
x = np.arange(-7, 7, 1)
y = np.arange(-7, 7, 1)
x, y = np.meshgrid(x, y)
z = 2*x + 3*y
ax.plot_surface(x, y, z, color='green', alpha=0.5)
ax.plot_surface(x, y, z + 40, color='blue', alpha=0.5)
plt.show()
```

5.2 Нарисуйте трехмерный график двух любых поверхностей второго порядка

```python
# Гиперболический параболоид
fig = figure()
ax = Axes3D(fig)
x = np.arange(-7, 7, 1)
y = np.arange(-7, 7, 1)
x, y = np.meshgrid(x, y)
z = x ** 2 - y ** 2
ax.plot_surface(x, y, z, alpha=0.8, color='green')
```

```python
# Конус
fig = figure()
ax = Axes3D(fig)
x = np.arange(-7, 7, 1)
y = np.arange(-7, 7, 1)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + y**2)
ax.set_xlim3d(-9, 9)
ax.set_ylim3d(-9, 9)
ax.set_zlim3d(-10, 10)
ax.plot_surface(x, y, z+0.5, alpha=0.8, edgecolor='gray', cmap=cm.Blues)
ax.plot_surface(x, y, -z-0.5, alpha=0.8, edgecolor='gray', cmap=cm.Blues_r)
```
