```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 51)
```

```python
plt.plot(x, np.cos(x), color="green")
plt.plot(x, np.cos(1.5*x), color="red")
plt.title("График cos(k*x)")
plt.xlabel("Значения x")
plt.ylabel("Значения y")
plt.axis([0.5, 10, -1.5, 1.5])
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, np.cos(x))
ax2.plot(x, np.cos(1.5*x))
ax1.set_title('График 1')
ax2.set_title('График 2')
ax1.grid()
ax2.grid()
fig.set_size_inches(16,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
