```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Задание 4

```python
a = np.linspace(0.01, 0.02, 100)
n = [1, 2, 3]
for i in n:
    x = i*np.pi/a
    plt.plot(a, x)
    plt.xlim(0.01, 0.02) 
    plt.ylim(100, 500)
    max_x=i*np.pi/0.01
    if max_x>500:
        max_x=500
    print(f"{i*np.pi/0.02} < x < {max_x}")
```
